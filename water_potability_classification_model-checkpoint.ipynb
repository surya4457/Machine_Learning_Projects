We will treat the outliers.

```python
l = [0,1,2,3,4,5,6,7,8]
def outlier(df):
    for i in df.iloc[:,l].columns:
        if df[i].dtype != 'str':
            Q1 = df[i].quantile(0.25)
            Q3 = df[i].quantile(0.75)
            IQR = Q3-Q1
            lower = Q1-(1.5*IQR)
            upper = Q3+(1.5*IQR)
            df.loc[:,i] = np.where(df[i].values>upper, upper,df[i].values)
            df.loc[:,i] = np.where(df[i].values<lower, lower,df[i].values)
    return df

    
df1 = outlier(df)
df1.info()
        
```
